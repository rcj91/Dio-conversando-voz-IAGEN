1. Gravação de Áudio Com Python (e Uma Pitada de JavaScript) 🎤

```python
!pip install sounddevice scipy ipython
```

```python
import os
import sounddevice as sd
from scipy.io.wavfile import write
from IPython.display import Audio, display

def record(sec=5, fs=44100, file_name="request_audio_teste.wav"):
    print("Ouvindo...\n")
    
    # grava 1 canal (mono)
    audio = sd.rec(int(sec * fs), samplerate=fs, channels=1, dtype="int16")
    sd.wait()
    
    write(file_name, fs, audio)
    return os.path.abspath(file_name)

record_file = record(sec=5)

print("Arquivo salvo em:", record_file)
display(Audio(record_file, autoplay=False))
```

```python
import os

print("record_file =", record_file)
print("arquivo existe?", os.path.exists(record_file))
```

```python
import whisper
# Selecione o modelo do Whisper que melhor atenda às suas necessidades:
# https://github.com/openai/whisper#available-models-and-languages
model = whisper.load_model("small")

# Transcreve o audio gravado anteriormente.
result = model.transcribe(record_file, fp16=False, language=language)
transcription = result["text"]
print(transcription)
```
